```python
import requests
import time


api_url = 'http://api.open-notify.org/iss-now.json'
num_dp = 100
dp = []

for _ in range(num_dp):
    response = requests.get(api_url)
    if response.status_code == 200:
        data = response.json()
        timestamp = data['timestamp']
        latitude = float(data['iss_position']['latitude'])
        longitude = float(data['iss_position']['longitude'])
        dp.append({'timestamp': timestamp, 'latitude': latitude, 'longitude': longitude})
        print(f"Timestamp: {timestamp}, Latitude: {latitude}, Longitude: {longitude}")
    else:
        print(f"Error: Unable to fetch data (Status Code: {response.status_code})")
    time.sleep(10)

print("Data collection complete!")
```

```python
import matplotlib.pyplot as plt

times = [point['timestamp'] for point in dp]
lats = [point['latitude'] for point in dp]
longs = [point['longitude'] for point in dp]

plt.figure(figsize=(15, 8))
plt.plot(times, lats, label='Latitude')
plt.plot(times, longs, label='Longitude')
plt.xlabel('Timestamp')
plt.ylabel('Coordinates')
plt.title('Tracking the ISS')
plt.legend()
plt.grid(True)
plt.show()
```
